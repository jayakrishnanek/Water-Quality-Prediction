**Models Used :** Decision Tree Classifier,GaussianNB,MLPClassifier.

```python
#Data acquisition
import pandas as pd
data=pd.read_csv("waterQuality1.csv")
print(data)
```

```python
#starting data preprocesing
data.head()
```

```python
#droping empty value containing rows
data=data.dropna()
print(data)
```

```python
#finding data types to find data available in numeric form
data.dtypes
```

```python
#there is some NAN values found in the data set in the places of some number
#so we have to clean it

data['ammonia'].fillna(value=0,inplace=True)
```

```python
data.dtypes
```

```python
#After cleaning we have to change the datatypes
data['ammonia']=data['ammonia'].astype('float64')
data['is_safe']=data['is_safe'].astype('float64')
```

```python
#making sure there is no empty coloums
data.isnull().sum()
```

```python
#piechart
import matplotlib.pyplot as plt

plt.title("Outcome Using PIECHART")
x=data['is_safe'].value_counts()[1]
y=data['is_safe'].value_counts()[0]
pie=[268,500]
lab=("not safe"," safe")

plt.pie(pie,labels=lab,autopct='%0.1f')
plt.show()
```

```python
data.info()
```

```python
#Decision Tree Classifier
cols=["aluminium","ammonia","arsenic","barium","cadmium","chloramine","chromium","copper","flouride","bacteria","lead","nitrates","nitrites","mercury","perchlorate","radium","selenium","silver","uranium"]
x=data[cols]
y=data.is_safe
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.1,random_state=1)
#4. Training
# Import Decision Tree Classifier

from sklearn.tree import DecisionTreeClassifier

#create decision tree classifier object
clf=DecisionTreeClassifier() #making a copy of the alg to clf

#train decision tree classifier

abc=clf.fit(x_train,y_train)
#5. testing
#predict the response for test dataset

y_pred =abc.predict(x_test)

from sklearn import metrics
print("accuracy",metrics.accuracy_score(y_test,y_pred)*100)

from sklearn import tree
tree.plot_tree(clf)

from sklearn.metrics import confusion_matrix
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score
conf_matrix =metrics.confusion_matrix(y_true=y_test, y_pred=y_pred)
acc=metrics.accuracy_score(y_test,y_pred)*100
pre=metrics.precision_score(y_test,y_pred)*100
F1=metrics.f1_score(y_test,y_pred)*100
rec=metrics.recall_score(y_test,y_pred)*100
print("Accuracy:",acc)
print("Precision:",pre)

print("Recall:",rec)
print("F1 Measure:",F1)
```

```python
#guassian nb
cols=["aluminium","ammonia","arsenic","barium","cadmium","chloramine","chromium","copper","flouride","bacteria","lead","nitrates","nitrites","mercury","perchlorate","radium","selenium","silver","uranium"]
x=data[cols]
y=data.is_safe
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.1,random_state=1)
#4. Training
# Import guassian nb

from sklearn.naive_bayes import GaussianNB

clf=GaussianNB() #making a copy of the alg to clf

#train algorithm

clf=clf.fit(x_train,y_train)

#5. testing
#predict the response for test dataset

y_pred =clf.predict(x_test)

from sklearn.metrics import confusion_matrix
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score
conf_matrix =metrics.confusion_matrix(y_true=y_test, y_pred=y_pred)
acc=metrics.accuracy_score(y_test,y_pred)*100
pre=metrics.precision_score(y_test,y_pred)*100
F1=metrics.f1_score(y_test,y_pred)*100
rec=metrics.recall_score(y_test,y_pred)*100
print("Accuracy:",acc)
print("Precision:",pre)

print("Recall:",rec)
print("F1 Measure:",F1)
```

```python
#ANN
cols=["aluminium","ammonia","arsenic","barium","cadmium","chloramine","chromium","copper","flouride","bacteria","lead","nitrates","nitrites","mercury","perchlorate","radium","selenium","silver","uranium"]
x=data[cols]
y=data.is_safe
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.1,random_state=1)
#4. Training
# Import ann

from sklearn.neural_network import MLPClassifier

#create ann object
clf=MLPClassifier(hidden_layer_sizes=(10,10,10),max_iter=500) #making a copy of the alg to clf

clf=clf.fit(x_train,y_train)

#5. testing
#predict the response for test dataset

y_pred =clf.predict(x_test)

from sklearn.metrics import confusion_matrix
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score
conf_matrix = confusion_matrix(y_true=y_test, y_pred=y_pred)
acc=metrics.accuracy_score(y_test,y_pred)*100
pre=metrics.precision_score(y_test,y_pred)*100
F1=metrics.f1_score(y_test,y_pred)*100
rec=metrics.recall_score(y_test,y_pred)*100
print("Accuracy:",acc)
print("Precision:",pre)

print("Recall:",rec)
print("F1 Measure:",F1)
```
